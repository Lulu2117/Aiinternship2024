```python
my_image = Image.open("otter.jpeg")
my_image.thumbnail((250,250))

my_image
```

```python
# Pixel average
# DONE
def pixel_average(p):
  return int((p[0]+p[1]+p[2])/3)
```

```python
# Replace bright pixels by grey pixels
# DONE
def replace_bright_pixels(image):
  for x in range(0, w):
    for y in range(0, h):
      pixel = my_image.getpixel((x,y))
      if pixel_average(pixel) > 153:
        image.putpixel((x,y), (pixel_average(pixel), pixel_average(pixel), pixel_average(pixel)))

replace_bright_pixels(my_image)
my_image
```

```python
my_image = Image.open("otter.jpeg")
my_image.thumbnail((250,250))

my_image
```

```python
#Replace pixels with R<120 to pink
# DONE
def replace_pink(image):
  for x in range(0, w):
    for y in range(0, h):
      pixel = my_image.getpixel((x,y))
      if pixel[0] < 120:
        image.putpixel((x,y), (255, 192, 203))

replace_pink(my_image)
my_image
```

```python
#Add a black border of x pixels: 2x+w and 2x+h
# DONE
my_image = Image.open("otter.jpeg")
my_image.thumbnail((250,250))

original_image = my_image

x = 20
y = 20

new_width = original_image.width + 2*x
new_height = original_image.height + 2*y
new_image = Image.new("RGB", (new_width, new_height), (0,0,0))
new_image.paste(original_image, (x,y))
new_image
```
